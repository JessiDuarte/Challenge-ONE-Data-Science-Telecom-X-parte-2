```python
import pandas as pd

df_final = pd.read_csv("/content/df_final.csv")

print(df_final.columns)
df_final.head()
```

```python

df_final = df_final.drop("IDCliente", axis=1)

df_final.head()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# -------------------------------
#  Definir las columnas categóricas
categorical_cols = [
    'Genero', 'AdultoMayor', 'Pareja', 'TipoContrato', 'FacturacionSinPapel',
    'MetodoPago', 'ServicioInternet', 'SeguridadOnline', 'RespaldoOnline',
    'ProteccionDispositivo', 'SoporteTecnico', 'StreamingTV',
    'StreamingPeliculas', 'ServicioTelefono', 'LineasMultiples'
]

# -------------------------------
#  Aplicar OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)  # drop='first' evita multicolinealidad
encoded_data = encoder.fit_transform(df_final[categorical_cols])

# Crear DataFrame de las columnas codificadas
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Combinar con columnas numéricas restantes
numerical_cols = df_final.drop(columns=categorical_cols).columns
df_final_encoded = pd.concat([df_final[numerical_cols].reset_index(drop=True), encoded_df], axis=1)

# -------------------------------
#  Revisar el resultado
print(df_final_encoded.columns.tolist())
df_final_encoded.head()
```

```python
# Conteo de cada clase
conteo_clases = df_final_encoded['Evasion'].value_counts()
print("Conteo de cada clase:\n", conteo_clases)

# Proporción de cada clase
proporcion_clases = df_final_encoded['Evasion'].value_counts(normalize=True)
print("\nProporción de cada clase:\n", proporcion_clases)

# Porcentajes
print(f"\nPorcentaje de clientes que cancelaron: {proporcion_clases[1]*100:.2f}%")
print(f"Porcentaje de clientes que permanecieron activos: {proporcion_clases[0]*100:.2f}%")
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# -------------------------------
#  Detectar columnas object
object_cols = df_final.select_dtypes(include='object').columns.tolist()
print("Columnas categóricas detectadas:", object_cols)

# -------------------------------
#  Convertir columnas object a numéricas
df_numeric = pd.get_dummies(df_final, columns=object_cols, drop_first=True)

# -------------------------------
#  Separar features y target
X = df_numeric.drop('Evasion', axis=1)
y = df_numeric['Evasion']

# -------------------------------
# Aplicar SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# -------------------------------
# Crear DataFrame balanceado
df_balanced = pd.concat([pd.DataFrame(X_res, columns=X.columns),
                         pd.DataFrame(y_res, columns=['Evasion'])], axis=1)

# -------------------------------
# Verificar proporciones
print(df_balanced['Evasion'].value_counts(normalize=True))
```

```python
from sklearn.preprocessing import StandardScaler

# Columnas numéricas a escalar (todas menos Evasion)
cols_to_scale = df_balanced.drop('Evasion', axis=1).columns

scaler = StandardScaler()
df_balanced[cols_to_scale] = scaler.fit_transform(df_balanced[cols_to_scale])

# Revisar resultados
df_balanced.head()
```

```python
from sklearn.model_selection import train_test_split

# Features y target
X = df_balanced.drop('Evasion', axis=1)
y = df_balanced['Evasion']

# División 70% entrenamiento / 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Verificar tamaños
print("Tamaño entrenamiento:", X_train.shape)
print("Tamaño prueba:", X_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# -------------------------------
#  Separar features y target
X = df_balanced.drop('Evasion', axis=1)
y = df_balanced['Evasion']

# -------------------------------
# Dividir en entrenamiento y prueba (70/30)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# -------------------------------
#  Escalado para Regresión Logística
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# -------------------------------
# Modelo 1: Regresión Logística (requiere normalización)
log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train_scaled, y_train)
y_pred_log = log_reg.predict(X_test_scaled)

print("=== Regresión Logística ===")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
print(confusion_matrix(y_test, y_pred_log))

# -------------------------------
# Modelo 2: Random Forest (no requiere normalización)
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

y_pred_log = log_reg.predict(X_test_scaled)

print("=== Regresión Logística ===")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print("Precision:", precision_score(y_test, y_pred_log))
print("Recall:", recall_score(y_test, y_pred_log))
print("F1-score:", f1_score(y_test, y_pred_log))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_log))
```

```python
y_pred_rf = rf.predict(X_test)

print("=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall:", recall_score(y_test, y_pred_rf))
print("F1-score:", f1_score(y_test, y_pred_rf))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_rf))
```

Análisis comparativo de modelos

Regresión Logística: Es un modelo interpretativo que requiere normalización. Presenta F1-score y recall moderados, lo que lo hace útil para explicar qué factores influyen en la cancelación de clientes. Puede presentar underfitting si no captura todas las tendencias de los datos; se puede ajustar incluyendo más variables o modificando la regularización.

Random Forest: Modelo más potente para predicción, capaz de capturar relaciones no lineales. Muestra F1-score y recall más altos, prediciendo de manera más efectiva las cancelaciones. Podría presentar overfitting si la diferencia entre accuracy de entrenamiento y prueba es muy alta; se puede ajustar reduciendo la profundidad máxima o aumentando el tamaño mínimo de las hojas.

Conclusión: Si priorizamos la precisión de predicción, Random Forest es la mejor opción. Si priorizamos la interpretación de qué variables influyen en la cancelación, la Regresión Logística es más adecuada.
